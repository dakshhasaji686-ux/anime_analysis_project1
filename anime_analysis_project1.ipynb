```python
def extract_ep(txt):
  data=""
  check=False
  for i in txt:
    if i==")":
      check=False
      return data
    elif check==True:
      data=data+i
    elif i=="(":
      check=True
```

```python
df['total_episode']=df['Title'].apply(extract_ep)
df
```

```python
df["total_episode"]=df["total_episode"].str.replace("eps","")
df['total_episode']=df['total_episode'].astype(int)
```

```python
df
```

```python
def extract_time(txt):
  data=""
  for i in range(len(txt)):
    if txt[i]==")":
      for j in range(i+1,i+20):
        data+=txt[j]
  return data
```

```python
df['time_period']=df['Title'].apply(extract_time)
df
```

```python
from datetime import datetime
def extract_duration(period):
  start,end=period.split("-")
  start_date=pd.to_datetime(start.strip(),format="%b %Y")
  end_date=pd.to_datetime(end.strip(),format="%b %Y")
  return abs((start_date.year-end_date.year)*12+(start_date.month-end_date.month))
```

```python
df['duration']=df['time_period'].apply(extract_duration)
df
```
